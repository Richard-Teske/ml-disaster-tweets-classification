```python
keywords_order = train[train["target"] == 1].groupby(["keyword"]).count()["target"].sort_values(ascending=False).keys()

g = sns.catplot(y="keyword", hue="target", data=train, kind="count", order=keywords_order)
g.fig.set_size_inches(20,50)
```

Few observations:

* location needs a pre process to make cities more generic
* keywords has special characters like %20 that needs to be removed, also apply stemming

```python
df = train.copy()
```

```python
stemmer = nltk.stem.porter.PorterStemmer()
```

```python
df["keyword"] = df["keyword"].apply(lambda x: str(x).replace("%20", " "))
df["keyword_stemmer"] = df["keyword"].apply(lambda x: stemmer.stem(str(x)))
```

```python
print("Original keyword length: {}".format(df["keyword"].unique().size))
print("Stemmer keyword length: {}".format(df["keyword_stemmer"].unique().size))
```

```python
df["location_split"] = df["location"].apply(lambda x: str(x).replace(" ", "").split(",")[-1])
df["location_split"].value_counts().head(20)
```

```python
print("Original location length: {}".format(df["location"].unique().size))
print("Split location length: {}".format(df["location_split"].unique().size))
```
