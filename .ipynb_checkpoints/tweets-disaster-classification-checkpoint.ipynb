Few observations:

* location needs a pre process to make cities more generic
* keywords has special characters like %20 that needs to be removed, also apply stemming

```python
df = train.copy()
```

```python
stemmer = nltk.stem.porter.PorterStemmer()
```

```python
df["keyword"] = df["keyword"].apply(lambda x: str(x).replace("%20", " "))
df["keyword_stemmer"] = df["keyword"].apply(lambda x: stemmer.stem(str(x)))
```

```python
print("Original keyword length: {}".format(df["keyword"].unique().size))
print("Stemmer keyword length: {}".format(df["keyword_stemmer"].unique().size))
```

```python
df["location_split"] = df["location"].apply(lambda x: str(x).replace(" ", "").split(",")[-1])
df["location_split"].value_counts().head(20)
```

```python
print("Original location length: {}".format(df["location"].unique().size))
print("Split location length: {}".format(df["location_split"].unique().size))
```

### GloVe

```python
from sklearn.manifold import TSNE
```

```python
embeddings = {}
with open("dataset/glove.6B.50d.txt", "r", encoding="utf-8") as file:
    for line in file:
        values = line.split()
        word = values[0]
        vector = np.array(values[1:], "float32")
        embeddings[word] = vector
```

```python
tsne = TSNE(n_components=2, random_state=0)
```

```python
words = list(embeddings.keys())
vectors = [embeddings[word] for word in words]
```

```python
glove_df = pd.DataFrame(vectors, index=words)
```

```python
glove_df = sklearn.utils.shuffle(glove_df)
glove_df = glove_df.head(200)
```

```python
%%time

X = tsne.fit_transform(glove_df)
```

```python
x = X[:, 0]
y = X[:, 1]
```

```python
fig, ax = plt.subplots(figsize=(20,12))
plt.scatter(x, y)

for i, txt in enumerate(glove_df.index):
    ax.annotate(txt, (x[i], y[i]))
    
plt.show()
```
